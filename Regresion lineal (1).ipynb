Ejercicio de Regresion Lineal 

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import pandas as pd
import numpy as np 
import os
```

```python
lineal = pd.read_csv(filepath_or_buffer = "C:\datacc\data.csv", sep=",")
```

```python
lineal.head(5)
```

```python
lineal.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
metro = lineal["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metro, lineal["precio"])
print("Interseccion (b)" , modelo.intercept_)
print("Pendiente (m)", modelo.coef_) 
```

```python
entrada = [[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print (predicciones)
```

```python
lineal.plot.scatter(x="metro", y="precio", label ="Datos Originales")
plt.scatter(entrada, predicciones, color="red")
plt.plot(entrada, predicciones, color="black", label="linea de regresion")
plt.xlabel("metro")
plt.ylabel("precio")
plt.legend()
plt.show()
```

El modelo de regresin lineal sugiere una relacin positiva entre el tamao en metros cuadrados y el precio de las propiedades. A medida que el tamao en metros cuadrados aumenta, se espera que el precio aumente de acuerdo con la pendiente positiva (5.65).
La interseccin (361.26) indica que incluso para propiedades con un tamao de 0 metros cuadrados, se espera un precio de 361.26 unidades monetarias.
Las predicciones realizadas por el modelo son coherentes con esta relacin, y se espera que el precio aumente a medida que el tamao en metros cuadrados aumenta.
La grfica de dispersin muestra que las observaciones originales siguen en gran medida la tendencia lineal sugerida por el modelo.
